```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#visualizing the appearance of a Keyword 'Samsung' in Youtube Search 

trend_data = pd.read_csv('SamsungTimeline.csv')
trend_df = pd.DataFrame(trend_data)

plt.figure(figsize=(13,3))

plt.plot(trend_df['Week'], trend_df['Score'])

#x-axis ticks - - data spacing
plt.xticks(trend_df['Week'][::5])


plt.xlabel("Weeks of 2023")
plt.ylabel("Search Scores")

title_font = {'fontsize': 12, 'color': 'blue'}
plt.title("The Keyword 'Samsung' appearance in Youtube Search", title_font)
plt.show()
```

```python
#visualizing and comparing the appearance of a Keyword 'Samsung' 'Innovation' 'Campus' in Youtube Search 

trend_data = pd.read_csv('comparison.csv')
df = pd.DataFrame(trend_data)

x = df['Samsung']
y = df['Innovation']
z = df['Campus']
t = df['Week']

#aspect ratio
plt.figure(figsize=(13,3))

plt.xlabel("Weeks of 2023")
plt.ylabel("Search Scores")

plt.plot(t, x, label='Samsung') #label for legend -- index table
plt.plot(t,y, label='Innovation')
plt.plot(t,z, label='Campus')

#spacing for t-data
plt.xticks(t[::5])


title_font = {'fontsize': 12, 'color': 'brown'}
plt.title("The Keyword 'Samsung' 'Innovation' 'Campus' appearance in Youtube Search", title_font)

plt.legend() #legend -- table

plt.show()
```
